### Simulación desnormalizada en python:

```python
# Módulos para Python, algo parecido a los #include de C

# módulos numéricos y de funciones científicas
import numpy as np
from scipy import signal as sig

# PyTC2: La librería para TC2
from pytc2.sistemas_lineales import bodePlot, pzmap, GroupDelay, analyze_sys
```

```python

import matplotlib as mpl
from matplotlib import pyplot as plt
#%%  Inicialización de librerías
# Setup inline graphics: Esto lo hacemos para que el tamaño de la salida, 
# sea un poco más adecuada al tamaño del documento

plt.figure(1)
plt.close(1)

#%% Esto tiene que ver con cuestiones de presentación de los gráficos,
# Podés modificar el ancho y alto de los gráficos que generás
fig_sz_x = 13
fig_sz_y = 7
fig_dpi = 80 # dpi
fig_font_size = 11

plt.rcParams.update({'font.size':fig_font_size})

# módulos numéricos y de funciones científicas
import numpy as np
from scipy import signal as sig

# Ahora importamos las funciones de PyTC2
from pytc2.sistemas_lineales import bodePlot, pzmap, GroupDelay, analyze_sys

from pytc2.general import print_subtitle
```

```python
# Definimos los parámetros de la simulación

#################################
R1=15915                    # Ohms
R2=11253                    # Ohms
R3=15915                    # Ohms
K = -1
Q = R2/R3
f0 = 10000                   # Hz
wp = 2 * np.pi * f0          # rad/s
w0_square = (R3/R1)*(wp**2)          # rad/s
######################################################
# Coeficientes de la transferencia desnormalizada
num = [K * w0_square]               # K*w0^2
den = [1, wp/Q, wp**2]       # s^2 + wp/Q s + wp^2

H_desnormalizada = sig.TransferFunction(num, den)

# Frecuencia personalizada (log spaced)
f = np.logspace(2, 6, 1000)           # De 10^2 a 10^6 Hz
w = 2 * np.pi * f                     # En rad/s

# Obtenemos la respuesta en frecuencia
w, mag, phase_deg = sig.bode(H_desnormalizada, w)
phase_rad = np.deg2rad(phase_deg)    # Convertimos fase a radianes

# Graficamos
plt.figure(figsize=(10, 6))

# Módulo
plt.subplot(2, 1, 1)
plt.semilogx(f, mag, label='Módulo [dB]')
plt.axvline(x=f0, color='r', linestyle='--', label='f₀ = 10kHz')
plt.title('Bode Plot - Transferencia Desnormalizada')
plt.ylabel('Magnitud [dB]')
plt.ylim([-60, 30])
plt.xlim([1e2, 1e6])
plt.grid(True, which='both')
plt.legend()

# Fase
plt.subplot(2, 1, 2)
plt.semilogx(f, phase_rad, label='Fase [rad]')
plt.axvline(x=f0, color='r', linestyle='--')
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Fase [rad]')
plt.ylim([-2, 4])
plt.xlim([1e2, 1e6])
plt.grid(True, which='both')
plt.legend()

# Diagrama de polos y ceros
pzmap(H_desnormalizada)

plt.tight_layout()
plt.show()
```

### Tipo de Filtro: Pasabajos
#### En caso que la ganancia atenúe mucho y el Q tenga un valor muy alto, el filtro podría funcionar como pasabanda
#### Ej: Wp=5*Wo, Q=25:

```python
# Definimos los parámetros de la simulación para simular pasabanda:

#################################
R3=15915                    # R3/R1 = 1/25
R1=25*R3                    # ohms
f0 = 10000                  # Hz
wp = 2 * np.pi * f0         # rad/s
Q=25
R2=Q*R3                  # Q=25 , 397 Kohm
K = -1                  # 
w0_square = (R3/R1)*(wp**2) #  w0^2= 1/25 wp^2 -> w0= 1/5 wp
######################################################
# Coeficientes de la transferencia desnormalizada
num = [K * w0_square]               # K*w0^2
den = [1, wp/Q, wp**2]       # s^2 + wp/Q s + wp^2

H_desnormalizada = sig.TransferFunction(num, den)

# Frecuencia personalizada (log spaced)
f = np.logspace(2, 6, 1000)           # De 10^2 a 10^6 Hz
w = 2 * np.pi * f                     # En rad/s

# Obtenemos la respuesta en frecuencia
w, mag, phase_deg = sig.bode(H_desnormalizada, w)
phase_rad = np.deg2rad(phase_deg)    # Convertimos fase a radianes

# Graficamos
plt.figure(figsize=(10, 6))

# Módulo
plt.subplot(2, 1, 1)
plt.semilogx(f, mag, label='Módulo [dB]')
plt.axvline(x=f0, color='r', linestyle='--', label='f₀ = 10kHz')
plt.title('Bode Plot - Transferencia Desnormalizada')
plt.ylabel('Magnitud [dB]')
plt.ylim([-60, 30])
plt.xlim([1e2, 1e6])
plt.grid(True, which='both')
plt.legend()

# Fase
plt.subplot(2, 1, 2)
plt.semilogx(f, phase_rad, label='Fase [rad]')
plt.axvline(x=f0, color='r', linestyle='--')
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Fase [rad]')
plt.ylim([-2, 4])
plt.xlim([1e2, 1e6])
plt.grid(True, which='both')
plt.legend()

# Diagrama de polos y ceros
pzmap(H_desnormalizada)

plt.tight_layout()
plt.show()
```
