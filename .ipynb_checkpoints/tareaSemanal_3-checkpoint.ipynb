### Recapitulando en limpio:

\begin{equation}
Siendo:
{\omega_p}^2=\frac{1}{C^2R3^2} ,  \frac{\omega_p}{Q} = \frac{1}{CR2} ,
K=-1 ,{\omega_o}^2=\frac{1}{C^2R1R3} , Q=\frac{R2}{R3}
\end{equation}

\begin{equation}
T(S) = K\frac{{\omega_o}^2}{{S}^2+{S}\frac{\omega_p}{Q}+{\omega_p}^2}
\end{equation}

### Con normas de impedancia y pulsación angular:
\begin{equation}
Normalizo: \ohm_z=R3, \ohm_w=\omega_p
\end{equation}
\begin{equation}
Siendo:
Q=R2^*,{\omega_p}^2=\frac{1}{{C^*}^2} ,{\omega_o}^2=\frac{1}{{C^*}^2{R1}^*},entonces -> {\omega_o}^2=\frac{1}{{R1}^*}{\omega_p}^2
\end{equation}

\begin{equation}
Redefino:
K=\frac{-1}{R1^*}
\end{equation}

\begin{equation}
T({S^*}) = K\frac{{1}}{{S^*}^2+{S^*}\frac{1}{Q}+1}
\end{equation}

### Simulación desnormalizada en python:

```python
# Módulos para Python, algo parecido a los #include de C

# módulos numéricos y de funciones científicas
import numpy as np
from scipy import signal as sig

# PyTC2: La librería para TC2
from pytc2.sistemas_lineales import bodePlot, pzmap, GroupDelay, analyze_sys
```

```python

import matplotlib as mpl
from matplotlib import pyplot as plt
#%%  Inicialización de librerías
# Setup inline graphics: Esto lo hacemos para que el tamaño de la salida, 
# sea un poco más adecuada al tamaño del documento

plt.figure(1)
plt.close(1)

#%% Esto tiene que ver con cuestiones de presentación de los gráficos,
# Podés modificar el ancho y alto de los gráficos que generás
fig_sz_x = 13
fig_sz_y = 7
fig_dpi = 80 # dpi
fig_font_size = 11

plt.rcParams.update({'font.size':fig_font_size})

# módulos numéricos y de funciones científicas
import numpy as np
from scipy import signal as sig

# Ahora importamos las funciones de PyTC2
from pytc2.sistemas_lineales import bodePlot, pzmap, GroupDelay, analyze_sys

from pytc2.general import print_subtitle
```

```python
# Definimos los parámetros de la simulación

#################################
K = 10
Q = 20
f0 = 10000                   # Hz
w0 = 2 * np.pi * f0          # rad/s
######################################################
# Coeficientes de la transferencia desnormalizada
num = [K * w0/Q, 0]               # (K/Q)*s
den = [1, w0/Q, w0**2]       # s^2 + w0/Q s + w0^2

H_desnormalizada = sig.TransferFunction(num, den)

# Frecuencia personalizada (log spaced)
f = np.logspace(2, 6, 1000)           # De 10^2 a 10^6 Hz
w = 2 * np.pi * f                     # En rad/s

# Obtenemos la respuesta en frecuencia
w, mag, phase_deg = sig.bode(H_desnormalizada, w)
phase_rad = np.deg2rad(phase_deg)    # Convertimos fase a radianes

# Graficamos
plt.figure(figsize=(10, 6))

# Módulo
plt.subplot(2, 1, 1)
plt.semilogx(f, mag, label='Módulo [dB]')
plt.axvline(x=f0, color='r', linestyle='--', label='f₀ = 10kHz')
plt.title('Bode Plot - Transferencia Desnormalizada')
plt.ylabel('Magnitud [dB]')
plt.ylim([-60, 30])
plt.xlim([1e2, 1e6])
plt.grid(True, which='both')
plt.legend()

# Fase
plt.subplot(2, 1, 2)
plt.semilogx(f, phase_rad, label='Fase [rad]')
plt.axvline(x=f0, color='r', linestyle='--')
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Fase [rad]')
plt.ylim([-2, 2])
plt.xlim([1e2, 1e6])
plt.grid(True, which='both')
plt.legend()

# Diagrama de polos y ceros
pzmap(H_desnormalizada)

plt.tight_layout()
plt.show()
```
