### Aproximación Chebyshev:

$$
|T_{LPF}(j\omega)|^2 = \frac{1}{1+\epsilon^2 \cdot C_n^2(\omega)}
$$
#### donde:
$$
C_n^2(\omega) = cosh^2 \left(n \cdot cosh^{-1}(\omega) \right)
$$
#### Utilizaremos su forma polinómica en base a n:
$$
C_n(\omega) = 2 \cdot \omega \cdot C_{n-1}(\omega) - C_{n-2}(\omega)
$$
$$
C_0 = 1
$$
$$
C_1 = \omega
$$

#### Para hallar el n, verificaremos en base a la plantilla de forma iterativa, teniendo en cuenta que al ser una plantilla pasa altos y estando diseñando un pasabajos, tendremos que el ${\omega}_{s_{LPF}} = \frac{1}{{\omega}_{s_{HPF}}}$

#### Normalizando para $\omega_c = 1$ nos queda:
$$
\omega_c = 1
$$
$$
{\omega}_{s_{HPF}} = \frac{f_s}{f_c} = \frac{10KHz}{40KHz} = \frac{1}{4} \rightarrow {\omega}_{s_{LPF}} = 4
$$

$$
\alpha_{max}[dB] = 0.5 = \left. 10 \cdot log(1 + \epsilon^2) \right|_{\omega = 1} \rightarrow \epsilon^2 = 10^{\frac{\alpha_{max}[dB]}{10}} -1
$$

$$
\epsilon^2 = 10^{\frac{0.5}{10}} - 1 \approx 0.122
$$

$$
\alpha_{min}[dB] = 30 \leq 10 \cdot log \left( 1 + \epsilon^2 \cdot {cosh}^2 \left( n \cdot {cosh}^{-1}({\omega_{s_{LPF}}}) \right) \right)
$$

$$
n=2 \rightarrow 10 \cdot log \left( 1 + 0.122 \cdot {cosh}^2 \left( 2 \cdot {cosh}^{-1}(4) \right) \right) \approx 20.72dB \rightarrow no \ es \ suficiente.

$$
n=3 \rightarrow 10 \cdot log \left( 1 + 0.122 \cdot {cosh}^2 \left( 3 \cdot {cosh}^{-1}(4) \right) \right) \approx 38.612dB \rightarrow satisface\ la\ plantilla.
$$

$$
C_2 = 2 \cdot \omega^2 -1

$$

$$

$$
|T_{LPF}(j\omega)|^2 = \frac{1}{1+\epsilon^2 \cdot C_n^2(\omega)} = \frac{1}{1 + \epsilon^2 \cdot \left( 2 \cdot \omega \cdot \left( 2 \cdot \omega^2 -1 \right) - \omega \right)^2} = \left. T(\$) \cdot T(-\$) \right|_{\omega=\frac{\$}{j}}

$$

#### Importación de librerías a usar:

```python
import sympy as sp
from pytc2.remociones import remover_polo_dc
from pytc2.general import a_equal_b_latex_s, print_latex, s, symbfunc2tf, factorSOS
from pytc2.sistemas_lineales import bodePlot
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal as sig
import math
```

```python
# ── Parámetros ────────────────────────────────────────────────
w1          = 1 / 0.626
w2_cuadrado = 1 / 1.142
w2          = np.sqrt(w2_cuadrado)
Q2          = w2 / (0.626 / 1.412) # 2.1107

# ── Filtro 1: pasa-altos 1er orden  H1(s) = s / (s + w1) ─────
num1 = np.array([1., 0.])
den1 = np.array([1., w1])

# ── Filtro 2: pasa-altos 2do orden ───────────────────────────
#   H2(s) = s² / (s² + (w2/Q2)·s + w2²)
num2 = np.array([1., 0., 0.])
den2 = np.array([1., w2 / Q2, w2_cuadrado])

# ── Cascada: convolver numeradores y denominadores ────────────
num_total = np.convolve(num1, num2)   # producto de polinomios
den_total = np.convolve(den1, den2)

H_total = sig.TransferFunction(num_total, den_total)

print("Numerador total  :", num_total)
print("Denominador total:", den_total)

# ── Bode (módulo y fase) ──────────────────────────────────────
w = np.logspace(-2, 2, 1000)          # rango de frecuencias
w_out, H_mag, H_phase = sig.bode(H_total, w=w)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6), sharex=True)
fig.suptitle("Respuesta en frecuencia — Filtro pasa-altos en cascada")

ax1.semilogx(w_out, H_mag)
ax1.axvline(1.0,     color="red",    ls="--", lw=1.2, label="wc = 1 rad/s")
ax1.axvline(0.25, color="orange", ls="--", lw=1.2, label=f"ws = 1/4 rad/s")
ax1.set_ylabel("Módulo (dB)")
ax1.set_ylim([-50, 10])
ax1.grid(True, which="both")
ax1.legend()

ax2.semilogx(w_out, H_phase)
ax2.axvline(1.0,     color="red",    ls="--", lw=1.2)
ax2.axvline(0.25, color="orange", ls="--", lw=1.2)
ax2.set_ylabel("Fase (°)")
ax2.set_ylim([-360, 0])
ax2.set_xlabel("Frecuencia (rad/s)")
ax2.grid(True, which="both")

plt.tight_layout()

# ── Mapa de polos y ceros ─────────────────────────────────────
zeros = np.roots(num_total)
poles = np.roots(den_total)

print("Ceros:", zeros)
print("Polos:", poles)

fig2, ax = plt.subplots(figsize=(6, 6))
ax.axhline(0, color="k", lw=0.8)
ax.axvline(0, color="k", lw=0.8)

ax.plot(zeros.real, zeros.imag, "go", ms=10, label="Ceros")
ax.plot(poles.real, poles.imag, "rx", ms=10, mew=2, label="Polos")

# Círculo unitario de referencia (opcional)
theta = np.linspace(0, 2 * np.pi, 300)
ax.plot(np.cos(theta), np.sin(theta), "b--", lw=0.7, alpha=0.4)

ax.set_xlabel(r"$\sigma$")
ax.set_ylabel(r"$j\omega$")
ax.set_title("Mapa de polos y ceros")
ax.legend()
ax.grid(True)
ax.set_aspect("equal")

plt.tight_layout()
plt.show()

w1_lp = 0.626
w2_lp = 1.142
w2_Q2_lp = 0.626
# ── Filtro 1: pasa-bajos 1er orden  H1(s) = w1 / (s + w1) ─────
num1_lp = np.array([0., w1_lp])
den1_lp = np.array([1., w1_lp])

# ── Filtro 2: pasa-bajos 2do orden ───────────────────────────
#   H2(s) = w2² / (s² + (w2/Q2)·s + w2²)
num2_lp = np.array([0., 0., w2_lp])
den2_lp = np.array([1., w2_Q2_lp, w2_lp])

# ── Cascada: convolver numeradores y denominadores ────────────
num_total_lp = np.convolve(num1_lp, num2_lp)   # producto de polinomios
den_total_lp = np.convolve(den1_lp, den2_lp)

zeros_lp = np.roots(num_total_lp)
poles_lp = np.roots(den_total_lp)

fig2, ax = plt.subplots(figsize=(6, 6))
ax.axhline(0, color="k", lw=0.8)
ax.axvline(0, color="k", lw=0.8)

ax.plot(zeros_lp.real, zeros_lp.imag, "go", ms=10, label="Ceros")
ax.plot(poles_lp.real, poles_lp.imag, "rx", ms=10, mew=2, label="Polos")

# Círculo unitario de referencia (opcional)
theta = np.linspace(0, 2 * np.pi, 300)
ax.plot(np.cos(theta), np.sin(theta), "b--", lw=0.7, alpha=0.4)

ax.set_xlabel(r"$\sigma$")
ax.set_ylabel(r"$j\omega$")
ax.set_title("Mapa de polos y ceros del Pasa bajos plantilla para comparación:")
ax.legend()
ax.grid(True)
ax.set_aspect("equal")

plt.tight_layout()
plt.show()
```

```python
# ── Desnormalización a fc = 40 kHz ────────────────────────────
fs = 10e3
fc = 40e3                          # Hz
wc = 2 * np.pi * fc                # rad/s  → ωc ≈ 251 327 rad/s

def desnormalizar(poly, wc):
    """Escala un polinomio normalizado al ωc indicado."""
    n = len(poly) - 1
    return np.array([coef / (wc ** (n - k)) for k, coef in enumerate(poly)])

num_dn = desnormalizar(num_total, wc)
den_dn = desnormalizar(den_total, wc)

H_dn = sig.TransferFunction(num_dn, den_dn)

# ── Bode desnormalizado ───────────────────────────────────────
w_dn = np.logspace(3, 7, 1000) * 2 * np.pi   # ~1 kHz a 10 MHz en rad/s
w_out_dn, H_mag_dn, H_phase_dn = sig.bode(H_dn, w=w_dn)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6), sharex=True)
fig.suptitle(f"Filtro desnormalizado — fc = {fc/1e3:.0f} kHz")

ax1.semilogx(w_out_dn / (2*np.pi), H_mag_dn)   # eje en Hz
ax1.axvline(fc, color="red",    ls="--", lw=1.2, label=f"fc = {fc/1e3:.0f} kHz")
ax1.axvline(fs, color="orange", ls="--", lw=1.2, label=f"fs = {fs/1e3:.0f} kHz")
ax1.set_ylabel("Módulo (dB)")
ax1.set_ylim([-60, 5])
ax1.grid(True, which="both")
ax1.legend()

ax2.semilogx(w_out_dn / (2*np.pi), H_phase_dn)
ax2.axvline(fc, color="red",    ls="--", lw=1.2)
ax2.axvline(fs, color="orange", ls="--", lw=1.2)
ax2.set_ylabel("Fase (°)")
ax2.set_xlabel("Frecuencia (Hz)")
ax2.grid(True, which="both")

plt.tight_layout()

# ── Mapa de polos y ceros desnormalizado ──────────────────────
zeros_dn = np.roots(num_dn)
poles_dn = np.roots(den_dn)

print(f"\n── Desnormalizado (fc = {fc/1e3:.0f} kHz) ──")
print("Ceros:", zeros_dn)
print("Polos:", poles_dn)

fig3, ax = plt.subplots(figsize=(6, 6))
ax.axhline(0, color="k", lw=0.8)
ax.axvline(0, color="k", lw=0.8)
ax.plot(zeros_dn.real, zeros_dn.imag, "go", ms=10, label="Ceros")
ax.plot(poles_dn.real, poles_dn.imag, "rx", ms=10, mew=2, label="Polos")
ax.set_xlabel("Re")
ax.set_ylabel("Im")
ax.set_title(f"Polos y ceros desnormalizados — fc = {fc/1e3:.0f} kHz")
ax.legend()
ax.grid(True)
ax.set_aspect("equal")

plt.tight_layout()
plt.show()
```

```python
# Parámetros:
Y1, Y2, Y3, Y4, Y5 = sp.symbols("Y1, Y2, Y3, Y4, Y5")
V1, V2, V3, V4 = sp.symbols("V1, V2, V3, V4")
G1, G2, C1, C2, C3, wt, w0 = sp.symbols("G1, G2, C1, C2, C3, wt, w0", real = True, posistive = True) 
s = sp.symbols('S')

# Consideramos V1 = Vi, V2 = Vo, V3 = VR1, V4 = V- operacional
aa = sp.solve([ 
                V3*(Y5+Y3+Y4+Y1) - V1*Y5 - V2*Y3, 
                -V2*Y2 - V3*Y4
                ], 
                [V2, V3])
T = sp.simplify(aa[V2]/V1)
print_latex(a_equal_b_latex_s('T(S)', T))

T_s = T.subs({
    Y1: G1,
    Y2: G2,
    Y3: s*C1,
    Y4: s*C2,
    Y5: s*C3
})
print_latex(a_equal_b_latex_s('T(S)', T_s))
T_s = sp.together(T_s)
T_s = sp.factor(T_s)
T_s = sp.simplify(T_s)
print_latex(a_equal_b_latex_s('T(S)', T_s))
```

#### Estudio:
$$
S^{\omega_0}_{C1} = \frac{C1}{\omega_0} \cdot \frac{\delta \omega_0}{\delta C1} = \frac{C1}{\sqrt{\frac{G1G2}{C1C2}}} \cdot \frac{\delta}{\delta C1} \left( G1^{\frac{1}{2}} G2^{\frac{1}{2}} C1^{\frac{-1}{2}} C2^{\frac{-1}{2}} \right)

$$

$$
