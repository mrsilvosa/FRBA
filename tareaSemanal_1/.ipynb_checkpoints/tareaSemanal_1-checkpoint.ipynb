### Análisis nodal: método no sistemático
Operacional realimentado, consideraremos V(-) y V(+) como 1 única tensión Vx 

#### Nodo entrada no inversora:
\begin{gather}
iC1 = iR3 \\
iC1 = C1 \cdot \overset{\bullet}{X} \\
C1 \cdot \overset{\bullet}{X} = iR3 = \frac{VR3}{R3} = \frac{Vx}{R3} \\
C1 \cdot \overset{\bullet}{X} = iC1 = \frac{V1}{ZC3+ZR3} (analizando \, la \, malla) \\
\frac{V1}{ZC3+ZR3} = \frac{V1}{\frac{1}{S\cdot C1}+R3} = \frac{V1 \cdot S \cdot C1}{S \cdot C1 \cdot R3 + 1} \\
\frac{V1 \cdot S \cdot C1}{S \cdot C1 \cdot R3 + 1} = \frac{Vx}{R3} \\
\frac{V1 \cdot S \cdot C1 \cdot R3}{S \cdot C1 \cdot R3 + 1} = Vx
\end{gather}

#### Nodo entrada inversora:
$$
iR1 = iR2
$$
$$
\frac{\left(V1-Vx\right)}{R1} = \frac{\left(Vx-V2\right)}{R2}
$$
$$
\frac{V1}{R1} = Vx \cdot \left(\frac{R1+R2}{R1 \cdot R2}\right) - \frac{V2}{R2}
$$
#### Reemplazamos Vx:
$$
\frac{V1}{R1} = \frac{V1 \cdot S \cdot C1 \cdot R3}{S \cdot C1 \cdot R3 + 1} \cdot \left(\frac{R1+R2}{R1 \cdot R2}\right) - \frac{V2}{R2}
$$
$$
V1 \cdot \left(\frac{1}{R1}-\frac{S \cdot C1 \cdot R3 \cdot \left(R1+R2\right)}{\left(S \cdot C1 \cdot R3 + 1 \right) \cdot R2 \cdot R1} \right) = \frac{-V2}{R2}
$$
$$
T(S) = \frac{V2}{V1} = -R2 \cdot \left(\frac{1}{R1}-\frac{S \cdot C1 \cdot R3 \cdot \left(R1+R2\right)}{\left(S \cdot C1 \cdot R3 + 1 \right) \cdot R2 \cdot R1} \right)
$$
#### Simplificando la expresión:
$$
T(S) = \frac{S\cdot C1\cdot R3\cdot R1 - R2}{S\cdot C1\cdot R3\cdot R1 + R1}
$$
$$
T(S) = \frac{S-\frac{R2}{C1\cdot R3\cdot R1}}{S+\frac{R1}{C1\cdot R3\cdot R1}}
$$

### Pasado en limpio:

\begin{equation}
Siendo: \omega_p=\frac{1}{C1\cdot R3} \, ,  \, \omega_o = \omega_p \cdot \frac{R2}{R1}
\end{equation}, 

\begin{equation}
T(S) = \frac{S-\omega_0}{S+\omega_p}
\end{equation}

#### Análisis módulo:
\begin{gather}
T(j\omega) = \frac{j\omega-\omega_0}{j\omega+\omega_p} = -1 \cdot \frac{-j\omega+\omega_0}{j\omega+\omega_p} = |T(\omega)| \cdot e^{\theta_{(\omega)}} \\
-1 = |1| \cdot e^{180º} \\
\frac{-j\omega+\omega_0}{j\omega+\omega_p} \cdot \frac{\omega_p - j\omega}{\omega_p - j\omega} = \frac{\omega_o \cdot \omega_p - \omega_o \cdot j\omega - \omega_p \cdot j\omega - {\omega}^2}{\omega^2 + {\omega_p}^2} = -1 \cdot \frac{- \omega_o \cdot \omega_p + {\omega}^2 + j\left(\omega \cdot \omega_o \cdot + \omega_p \cdot \omega\right) }{\omega^2 + {\omega_p}^2} \\
-1 \cdot -1 = 1 \\
T(j\omega) = \frac{\left(- \omega_o \cdot \omega_p + {\omega}^2 \right) + j\left(\omega \cdot \omega_o \cdot + \omega_p \cdot \omega\right) }{\omega^2 + {\omega_p}^2} 
\end{gather}

\begin{equation}
|T(\omega)| = \sqrt{\frac{(\omega^2-\omega_o \cdot \omega_p)^2+(\omega \cdot (\omega_p+\omega_o))^2}{(\omega^2+{\omega_p}^2)^2}}
\end{equation}

#### Análisis fase:
\begin{gather}
\theta_{(\omega)} = +tg^{-1}\left(\frac{\left(\omega \cdot \omega_o \cdot + \omega_p \cdot \omega\right)}{(\omega^2-\omega_o \cdot \omega_p)}\right)-tg^{-1}\left(\frac{0}{\omega^2 + {\omega_p}^2}\right) \\
tg^{-1} \left( \frac{0}{X} \right) = 0º \, para \, todo \, X \, distinto \, de \, 0
\end{gather}

### 3 análisis posibles:
$$
\omega_o = \omega_p
$$
$$
\omega_o > \omega_p
$$
$$
\omega_o < \omega_p
$$

#### Nos concentraremos en el caso donde R2=R1:
$$
\omega_o = \omega_p
$$

\begin{equation}
T(S) = \frac{S-\omega_0}{S+\omega_0}
\end{equation}

$$
T(j\omega) = \frac{\left(- {\omega_o}^2 + {\omega}^2 \right) + j\left(2 \cdot \omega \cdot \omega_o \right) }{\omega^2 + {\omega_0}^2} 
$$

\begin{equation}
|T(\omega)| = \sqrt{\frac{(\omega^2-{\omega_o}^2)^2+(2 \cdot \omega \cdot \omega_0)^2}{(\omega^2+{\omega_0}^2)^2}} = 1
\end{equation}

$$
\theta_{(\omega)} = +tg^{-1}\left(\frac{\left(2 \cdot \omega \cdot \omega_o\right)}{(\omega^2-{\omega_o}^2)}\right)
$$

### Normas:

#### Norma impedancia propuesta:
$$
\ohm_z = R3
$$

$$
T(S) = \frac{S-\frac{R2}{C1\cdot R3\cdot R1}}{S+\frac{R1}{C1\cdot R3\cdot R1}} = \frac{S-\frac{1}{C1\cdot R3}}{S+\frac{1}{C1\cdot R3}}
$$

$$
T(S)_{\ohm_z = R3} = \frac{S-\frac{1}{C1\cdot \ohm_z}}{S+\frac{1}{C1\cdot \ohm_z}} = \frac{S-\frac{1}{C'}}{S+\frac{1}{C'}}
$$

#### Norma frecuencia propuesta:
$$
\ohm_w = \omega_0
$$
\begin{equation}
T(S)_{\ohm_w = \omega_0} = \frac{S-1}{S+1}
\end{equation}

```python
# Definimos los parámetros de la simulación

#################################

######################################################
wp = 1
wo = 1

# Cargamos la funcion transferencia como vectores de sus coeficientes. asumo R2=R1

num = np.array([ 1., -wo ])
den = np.array([ 1., wp  ])

H1 = sig.TransferFunction( num, den )

# Graficamos el diagrama de polos y ceros
# Graficamos la respuesta en frecuencia para el modulo y la fase.

_, axes_hdl = bodePlot(H1)

# para que se vea como uno intuye el módulo. Probar comentar las siguientes 2 líneas
plt.sca(axes_hdl[0])
plt.ylim([-1,1])

plt.gca

pzmap(H1)

plt.show()
```

### Verificación de la transferencia con SymPy:

```python
import sympy as sp
s, w0 = sp.symbols('s w0')
print("1_:")
T = (s - w0) / (s + w0)
sp.pprint(T)

print("2_:")
w = sp.symbols('w', real=True)
j = sp.I
Tjw = T.subs(s, j*w)
sp.pprint(Tjw)

print("3_:")
sp.simplify(Tjw)
sp.pprint(Tjw)

print("4_:")
sp.pprint(sp.re(Tjw))
print("5_:")
sp.pprint(sp.im(Tjw))
print("6_:")
sp.pprint(sp.simplify(sp.Abs(Tjw)))
print("7_:")
sp.pprint(sp.arg(Tjw))
print("8_:")
Tjw_simplificado = sp.simplify(Tjw * (w0 - j*w)/(w0- j*w))
sp.pprint(sp.expand(Tjw_simplificado))
```
